```python
import pandas as pd

import sys
sys.path.append("..")
import helpers
import test_helpers

from IPython.display import display, HTML
```

```python
bucket = 'nasa-eodc-scratch'
directory = 'NLDAS/netcdf/test_files'
```

```python
credentials = helpers.get_credentials()
s3fsfs = helpers.create_s3filesystem(credentials)
```

```python
test_files = s3fsfs.glob(f's3://{bucket}/{directory}/*_Tair.nc')
```

```python
test_files
```

# 0.01 degree x 0.01 degree timeseries

```python
# these 2 need to match, precision equaling the spatial size's decimal points
spatial_size = 0.01
precision = 2
s3_times = {}
open_and_load_times = []
```

```python
lat_range, lon_range = test_helpers.select_random_spatial_range(spatial_size, precision)
for test_file in test_files:
    test_info, s3_request_info = test_helpers.load_series(test_file, s3fsfs, lat_range, lon_range)
    open_and_load_times.append(test_info)
    s3_times[test_file.split('/')[-1]] = s3_request_info
```

```python
pd.DataFrame(open_and_load_times)
```
