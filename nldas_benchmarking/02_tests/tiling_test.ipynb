```python
import s3fs
import test_helpers

import xarray as xr
import os
from titiler.xarray.io import Reader
import pandas as pd
import random
```

```python
test_files = [f"s3://{f}" for f in test_helpers.list_test_files(s3fsfs)]
```

# Test Zoom Level 0 Tile

```python
open_and_load_times = []
zoom = 0
random_idx = random.randint(1,24)
```

```python
for test_file in test_files:
    test_info, _ = test_helpers.load_tile(test_file, zoom, random_idx)
    open_and_load_times.append(test_info)
```
